```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('../data/owid-covid-data.csv')
df.head()
```

```python
india_df = df[df['location'] == 'India']
india_df.head()
```

```python
india_df['date'] = pd.to_datetime(india_df['date'])
```

```python
final_df = india_df[(india_df['date'].dt.year >= 2020) & 
                    (india_df['date'].dt.year <= 2021)]
```

```python
final_df = final_df.reset_index(drop=True)
```

```python
final_df.head()
```

```python
final_df.shape
```

```python
final_df.columns
```

```python
final_df.isnull().sum()
```

```python
use_columns = [
    'date',
    'total_cases',
    'new_cases',
    'total_deaths',
    'new_deaths',
    'total_tests',
    'new_tests',
    'people_vaccinated',
    'people_fully_vaccinated',
    'population'
]
```

```python
final_df_clean = final_df[use_columns].copy()
```

```python
final_df_clean[['new_cases', 'new_deaths', 'new_tests']] = \
final_df_clean[['new_cases', 'new_deaths', 'new_tests']].fillna(0)
```

```python
final_df_clean[['total_cases', 'total_deaths', 'total_tests',
                'people_vaccinated', 'people_fully_vaccinated']] = \
final_df_clean[['total_cases', 'total_deaths', 'total_tests',
                'people_vaccinated', 'people_fully_vaccinated']].ffill()
```

```python
final_df_clean = final_df_clean.fillna(0)
```

```python
final_df_clean.info()
```

```python
final_df_clean.head()
```

```python
final_df_clean.tail()
```

```python
final_df_clean['year'] = final_df_clean['date'].dt.year
final_df_clean['month'] = final_df_clean['date'].dt.month
final_df_clean['month_name'] = final_df_clean['date'].dt.strftime('%B')
```

```python
final_df_clean['death_rate'] = (
    (final_df_clean['total_deaths'] / final_df_clean['total_cases']) * 100
)
```

```python
final_df_clean['daily_case_growth'] = (final_df_clean['new_cases'] / final_df_clean['total_cases']) * 100
```

```python
final_df_clean['vaccinated_percent'] = (
    final_df_clean['people_fully_vaccinated'] / final_df_clean['population']
) * 100
```

```python
max_new_cases = final_df_clean['new_cases'].max()

final_df_clean['peak_case_day'] = final_df_clean['new_cases'].apply(
    lambda x: 1 if x == max_new_cases else 0
)
```

```python
final_df_model = final_df_clean.copy()
```

```python
final_df_model.head()
```

```python
final_df_model.info()
```

```python
monthly_df = final_df_model.groupby(
    ['year', 'month', 'month_name'], as_index=False
).agg({
    'new_cases': 'sum',
    'new_deaths': 'sum',
    'total_cases': 'max',     # End of month cumulative
    'total_deaths': 'max',
    'vaccinated_percent': 'max',
})
```

```python
yearly_df = final_df_model.groupby(
    ['year'], as_index=False
).agg({
    'new_cases': 'sum',
    'new_deaths': 'sum',
    'total_cases': 'max',
    'total_deaths': 'max',
    'vaccinated_percent': 'max',
})
```

```python
def get_wave(date):
    if pd.Timestamp('2020-04-01') <= date <= pd.Timestamp('2020-11-30'):
        return 'Wave 1'
    elif pd.Timestamp('2021-03-01') <= date <= pd.Timestamp('2021-06-30'):
        return 'Wave 2'
    else:
        return 'No Wave'

final_df_model['wave'] = final_df_model['date'].apply(get_wave)
```

```python
wave_df = final_df_model.groupby(
    ['wave'], as_index=False
).agg({
    'new_cases': 'sum',
    'new_deaths': 'sum',
    'total_cases': 'max',
    'total_deaths': 'max',
})
```

```python
final_df_model.to_csv('../data/covid_india_daily.csv', index=False)
```

```python
monthly_df.to_csv('../data/covid_india_monthly.csv', index=False)
```

```python
yearly_df.to_csv('../data/covid_india_yearly.csv', index=False)
```

```python
wave_df.to_csv('../data/covid_india_waves.csv', index=False)
```
